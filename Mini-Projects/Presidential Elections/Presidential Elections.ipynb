### Discussion
In this project, the visualization was concerned about correlation of average income with elections attendance. The point is that there was a research conducted in the USA stating that people with higher income tend to vote more (https://www.weforum.org/agenda/2018/07/low-voter-turnout-increasing-household-income-may-help). Therefore, the main aim of this project was to check whether or not this is applicable to Kazakhstan.

The generated plot shows the dependence between average income in a particular region with attendance rate in that region. As it can be seen, the dependence is negative, i.e. people in regions with higher average salary tend to vote less. 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from IPython.core.display import HTML
import matplotlib.pyplot as plt


from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
```

```python
req = requests.get("https://ru.wikipedia.org/wiki/%D0%9F%D1%80%D0%B5%D0%B7%D0%B8%D0%B4%D0%B5%D0%BD%D1%82%D1%81%D0%BA%D0%B8%D0%B5_%D0%B2%D1%8B%D0%B1%D0%BE%D1%80%D1%8B_%D0%B2_%D0%9A%D0%B0%D0%B7%D0%B0%D1%85%D1%81%D1%82%D0%B0%D0%BD%D0%B5_(2019)")
page = req.text
soup = BeautifulSoup(page, 'html.parser')
tables = soup.find_all('table')
[t['class'] for t in tables]
```

```python
table_att = HTML(str(tables[5]))
table_att
```

```python
rows = [row for row in tables[5].find_all("tr")]
rows
```

```python
values = [value.get_text() for row in rows[1:] for value in row.find_all("td")]
values
```

```python
attendance = [float(val.replace("\xa0%\n","").replace(",",".")) for val in values[2::3]]
attendance
```

```python
ind_pop = [val.replace('\n', '') for val in values[0::3]]
ind_pop
```

```python
att_df = pd.DataFrame(data = attendance, index=ind_pop, columns = ['Attendace by region'])
att_df
```

```python
avg_income = [114100,156927, 135612, 321933, 157098,112275,181616,169122, 144777,157928,289593, 157300, 130291,128088, 227370,138760,287677]
income = pd.DataFrame(data = avg_income, index = att_df.index, columns = ['Average income'])
```

```python
income
```

```python
final_df = pd.merge(att_df,income, left_index = True, right_index=True)
final_df
```

```python
y = final_df['Attendace by region']
x = final_df['Average income']
plt.scatter(x, y)
plt.ylabel('Attendance by region in %')
plt.xlabel('Average income by region')
plt.title('Elections Attendance vs Average income (by region)')
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"g-")

plt.show()
```
